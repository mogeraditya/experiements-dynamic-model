```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib
plt.style.use('dark_background')
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
data_file_dir = "/home/adityamoger/Documents/GitHub/dynamic_model_of_cocktail_party_nightmare/JammingExperiment/Data/AnalysisData/collision_counts_w_jammers1.csv"
df = pd.read_csv(data_file_dir)
df["jammer_resolution_by_walls"] = [int(i)*4 if i!="random" else i for i in df["jammer_resolution"]]
# df["jammer_resolution_by_walls"] = df["jammer_resolution"] * 4
df
```

```python
plt.figure(figsize=(15,10))
sns.color_palette("rocket")
sns.boxplot(data=df, x="jammer_resolution_by_walls", y="value", hue="behaviour", palette = sns.color_palette("hls", 3), errcolor = "white")#, capsize = 0.1)
plt.xlabel("number of jammers")
plt.ylabel("number of collisions in 50 seconds")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
plt.figure(figsize=(10,7))
sns.barplot(data=df, x="behaviour", y="value", errcolor = "white", capsize = 0.1)
plt.xlabel("number of jammers")
plt.ylabel("number of collisions in 50 seconds")
```
